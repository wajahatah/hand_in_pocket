```python
import cv2
import pandas as pd
import os

def annotate_sides(csv_path, video_folder, output_csv):
    df = pd.read_csv(csv_path)

    # Safety checks
    required_cols = ["camera", "video", "frame", "desk_no"]
    for col in required_cols:
        if col not in df.columns:
            raise ValueError(f"Missing column: {col}")

    # Create new columns
    if "right" not in df.columns:
        df["right"] = 0
    if "left" not in df.columns:
        df["left"] = 0

    # Cache opened videos
    video_cache = {}

    for idx, row in df.iterrows():
        cam = row["camera"]
        vid = row["video"]
        frame_num = int(row["frame"])
        desk = row["desk_no"]

        video_name = f"{cam}_{vid}.mp4"
        video_path = os.path.join(video_folder, video_name)

        if not os.path.exists(video_path):
            print(f"[SKIP] Video not found: {video_path}")
            continue

        # Load or reuse video
        if video_name not in video_cache:
            cap = cv2.VideoCapture(video_path)
            if not cap.isOpened():
                print(f"[ERROR] Cannot open: {video_path}")
                continue
            video_cache[video_name] = cap
        else:
            cap = video_cache[video_name]

        # Seek to frame
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)
        ok, frame = cap.read()
        if not ok:
            print(f"[ERROR] Failed to read frame {frame_num} in {video_name}")
            continue

        # Display frame
        frame = cv2.resize(frame, (1280,720))
        cv2.imshow("Frame", frame)
        print(f"/nDesk {desk} | {video_name} | Frame {frame_num}")
        print("RIGHT (r), LEFT (l), BOTH (b), Quit (q)")

        key = cv2.waitKey(0)

        if key == ord('q'):
            break
        elif key == ord('p'):
            df.at[idx, "right"] = -1
            df.at[idx, "left"] = -1
        elif key == ord('r'):
            df.at[idx, "right"] = 1
            df.at[idx, "left"] = 0
        elif key == ord('l'):
            df.at[idx, "right"] = 0
            df.at[idx, "left"] = 1
        elif key == ord('b'):
            df.at[idx, "right"] = 1
            df.at[idx, "left"] = 1
        else:
            print("Invalid key, skipping...")
            continue

    cv2.destroyAllWindows()

    # Save output
    df.to_csv(output_csv, index=False)
    print(f"/nSaved updated CSV to: {output_csv}")


if __name__ == "__main__":
    annotate_sides(
        csv_path="C:/wajahat/hand_in_pocket/dataset/training3/right_left/itteration3_right_left_annotation.csv",
        video_folder="F:/Wajahat/hand_in_pocket/traiining/train",
        output_csv="C:/wajahat/hand_in_pocket/dataset/training3/right_left/itteration3_right_left_annotation_sides.csv"
    )
```

```python
import re
import pandas as pd
import sys
import os

def extract_videos_to_csv(log_file_path, output_csv_filename): #="extracted_video_names.csv"):
    """
    Reads a log file, extracts video names (c#_v#) only from lines 
    that match the '[SKIP] Video not found: ...' format, removes duplicates, 
    and saves the unique names to a CSV file.
    """
    
    # Check if the file exists
    if not os.path.exists(log_file_path):
        print(f" Error: File not found at path: {log_file_path}")
        return

    # 1. Define the Strict Line Pattern for Extraction
    # It now strictly requires the leading part "[SKIP] Video not found: "
    # and includes the full file path pattern you provided.
    # Capture Group (the part we want) is: (c\d+_v\d+)
    pattern = r"\[SKIP\] Video not found: .*\\(c\d+_v\d+)\.mp4"

    try:
        # Read the content of the log file
        with open(log_file_path, 'r') as f:
            log_text = f.read()

        # 2. Extract the Video Names
        # re.findall returns only the contents of the capture group ()
        video_names = re.findall(pattern, log_text)

        # 3. Filter for Unique Names and Sort
        unique_video_names = sorted(list(set(video_names)))

        # 4. Save to CSV using Pandas
        df = pd.DataFrame(unique_video_names, columns=['Video_Name'])
        df.to_csv(output_csv_filename, index=False)

        print("--- Extraction Summary ---")
        print(f" Extraction complete!")
        print(f"Total unique video names extracted: **{len(unique_video_names)}**")
        print(f"Results saved to file: **{output_csv_filename}**")
        
    except Exception as e:
        print(f" An error occurred during processing: {e}")

# --- Main execution block ---
if __name__ == "__main__":
    # The script expects the file path as the first argument after the script name.
    if len(sys.argv) < 2:
        print("\nUsage Example:")
        print("python script_name.py **path/to/your/log_file.txt**")
    else:
        # sys.argv[1] is the file path provided by the user
        # input_file = sys.argv[1]
        input_file = "C:/Users/LT/Downloads/right_left_distribution.txt"
        output_file = "C:/Users/LT/Downloads/missing_files.csv"
        extract_videos_to_csv(input_file, output_file)
```
