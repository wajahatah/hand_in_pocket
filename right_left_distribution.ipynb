```python
import pandas as pd
import os 

def extract_positive_rows(input_csv, output_csv, output_folder):
    # Read the CSV
    df = pd.read_csv(input_csv)

    # Check column existence
    if "hand_in_pocket" not in df.columns:
        raise ValueError("Column 'hand_in_pocket' not found in the CSV.")

    # Filter rows where hand_in_pocket == 1
    df_filtered = df[df["hand_in_pocket"] == 1]

    # Save
    os.makedirs(output_folder, exist_ok=True)
    output_csv= os.path.join(output_folder,output_csv)
    df_filtered.to_csv(output_csv, index=False)

if __name__ == "__main__":

    input_csv = "C:/wajahat/hand_in_pocket/dataset/training3/itteration3_temp_norm_balanced.csv"
    output_folder = "C:/wajahat/hand_in_pocket/dataset/training3/right_left/" 
    output_csv = "itteration3_right_left_annotation.csv"
    extract_positive_rows(input_csv, output_csv, output_folder)
```
